K-Means Clustering

importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

importing the dataset

```python
dataset = pd.read_csv('/content/Mall_Customers.csv')
x= dataset.iloc[:,[3,4]].values
y= dataset.iloc[:,:-1].values
```

Using the elbow method to optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i,init = 'k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('THE Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

Training  the K-means model on the dataset

```python
kmeans = KMeans(n_clusters=5,init = 'k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
print(y_kmeans)
```

Visulaising the clusters

```python
plt.scatter(x[y_kmeans == 0 , 0],x[y_kmeans == 0 ,1], s = 100 , c = 'red', label= 'medium income and medium spending')
plt.scatter(x[y_kmeans == 1 , 0],x[y_kmeans == 1 ,1], s = 100 , c = 'blue', label= 'high income and low spending')
plt.scatter(x[y_kmeans == 2 , 0],x[y_kmeans == 2 ,1], s = 100 , c = 'cyan', label= 'low income and low spending')
plt.scatter(x[y_kmeans == 3 , 0],x[y_kmeans == 3 ,1], s = 100 , c = 'magenta', label= 'low income and high spending')
plt.scatter(x[y_kmeans == 4 , 0],x[y_kmeans == 4,1], s = 100 , c = 'green', label= 'high income and high spending')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s=300 , c ='yellow',label ='centroids')

plt.title('Cluster of customers')
plt.xlabel('Annual Income ($)')
plt.ylabel('spending score(1-100)')

plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))

plt.show()
```
