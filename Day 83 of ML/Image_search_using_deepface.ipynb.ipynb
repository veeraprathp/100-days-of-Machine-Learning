```python
import os
import cv2
import numpy as np
from google.colab import files
from deepface import DeepFace
```

```python
import os

# Define the dataset directory
dataset_directory = '/kaggle/input/pins-face-recognition/105_classes_pins_dataset/pins_Adriana Lima'

# List all files in the directory
files = os.listdir(dataset_directory)
print("Files in the dataset directory:")
for file in files:
    print(file)
```

```python
def get_file_paths(directory, extensions=['.jpg', '.png']):
    """
    Get all image file paths in the specified directory.

    Args:
    - directory (str): The directory path.
    - extensions (list of str): List of file extensions to include.

    Returns:
    - file_paths (list of str): List of file paths.
    """
    file_paths = []
    for file in os.listdir(directory):
        if any(file.endswith(ext) for ext in extensions):
            file_paths.append(os.path.join(directory, file))
    return file_paths
```

```python
# Define the dataset directory
dataset_directory = '/kaggle/input/pins-face-recognition/105_classes_pins_dataset/pins_Adriana Lima'

# Get image file paths
image_file_paths = get_file_paths(dataset_directory)

print("Number of image files found:", len(image_file_paths))
print("Sample image file paths:")
for path in image_file_paths[:5]:  # Print first 5 paths as an example
    print("Path:", path)

# Load images from the dataset
loaded_images = []
for file_path in image_file_paths:
    img = cv2.imread(file_path)
    if img is not None:
        loaded_images.append(img)

print(f"Loaded {len(loaded_images)} images from the dataset.")
```

```python
# Upload query images for comparison
uploaded_query_files = files.upload()
query_images = []

# Load uploaded query images into a list
for filename in uploaded_query_files.keys():
    img = cv2.imread(filename)
    if img is not None:
        query_images.append(img)

print(f"Uploaded {len(query_images)} query images.")
```

```python
# Function to extract embeddings
def extract_embeddings(images):
    embeddings = []
    for img in images:
        try:
            # Perform face embedding extraction with MTCNN
            embedding = DeepFace.represent(img, model_name='Facenet', detector_backend='mtcnn', enforce_detection=False)
            embeddings.append(embedding)
        except Exception as e:
            print(f"Error processing image: {e}")
    return embeddings

# Extract embeddings from loaded dataset images
loaded_embeddings = extract_embeddings(loaded_images)

# Extract embeddings from uploaded query images
query_embeddings = extract_embeddings(query_images)
```

```python
# Verify faces and display results
results = []
for query_embedding in query_embeddings:
    for loaded_embedding in loaded_embeddings:
        result = DeepFace.verify(query_embedding[0]['embedding'], loaded_embedding[0]['embedding'], model_name='Facenet', detector_backend='mtcnn')
        results.append(result)

# Display verification results
for idx, result in enumerate(results):
    print(f"Verification result for query image {idx + 1}: {result}")
```

```python
import os

# Directory to save matched images
matched_dir = "matched_faces"
os.makedirs(matched_dir, exist_ok=True)  # Create the directory if it doesn't exist

# Verify faces and display results
results = []
for query_idx, query_embedding in enumerate(query_embeddings):
    for loaded_idx, loaded_embedding in enumerate(loaded_embeddings):
        # Perform verification
        result = DeepFace.verify(query_embedding[0]['embedding'], loaded_embedding[0]['embedding'], model_name='Facenet', detector_backend='mtcnn')

        # Check if the result indicates a match
        if result['verified']:
            # Construct the file path for saving the matched image
            matched_image_path = image_file_paths[loaded_idx]  # Get the original file path of the matched image
            matched_image = cv2.imread(matched_image_path)  # Read the matched image
            query_image_name = f"query_image_{query_idx + 1}.jpg"  # Create a name for the saved query image

            # Save the matched image to the new directory
            cv2.imwrite(os.path.join(matched_dir, f"{query_image_name}_matched_{loaded_idx + 1}.jpg"), matched_image)
            results.append((query_image_name, matched_image_path, result))

# Display verification results
for idx, (query_name, matched_path, result) in enumerate(results):
    print(f"Verification result for {query_name}: {result}, Matched Image Path: {matched_path}")
```
